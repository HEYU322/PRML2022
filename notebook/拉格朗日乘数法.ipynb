在限制条件下的求极值，一般就是逆表示，再带入原方程.但太难算了<p>
$ g(x_1,x_2)=0 \tag{E.1}$

```python
# preparation
import numpy as np
import matplotlib.pyplot as plt
```

采用拉格朗日乘数法.首先D维向量x:$x_1,...,x_D$,限制方程$g(x)=0$，表示一个(D-1)维曲面。


```python
# 等式限制
fig, ax = plt.subplots(figsize=(10, 10))
d = plt.imread('./graph/等式限制.png')
plt.axis('off')
ax.imshow(d)
```

**限制函数的梯度$\triangledown g(x)$都正交于限制曲面**
,因为:限制曲面上点x和附近的点$x+\epsilon$.泰勒展开有：<p>
$$
g(\boldsymbol{x}+\boldsymbol{\epsilon}) \simeq g(\boldsymbol{x})+\boldsymbol{\epsilon}^{T} \nabla g(\boldsymbol{x}) \tag{E.2}
$$<p>
因为都在限制曲面上，g(x)和$g(x+\epsilon)$都是0。最后一项近似为零，再加上$\epsilon$很小，平行于限制曲面，所以$\nabla g(x)$都正交于限制曲面.

$\nabla f(x)$也正交于限制曲面。因为不满足的话，总有个方向可以使得f增大。

```python
# 自己用图
fig, ax = plt.subplots(figsize=(10, 10))
d = plt.imread('./graph/等式自己.png')
plt.axis('off')
ax.imshow(d)
```

所以有拉格朗日乘数$\lambda$：<p>
$\nabla f + \lambda \nabla g =0 \tag{E.3}$
拉格朗日函数：<p>
$L(x,\lambda)=f(x) + \lambda g(x) =0 \tag{E.4}$<p>
    y的极值问题转化为L的极值问题。限制问题转化为无限制问题。

```python
# 例子

fig, ax = plt.subplots(figsize=(10, 10))
d = plt.imread('./graph/例子.png')
plt.axis('off')
ax.imshow(d)
```

有$f(x_1,x_2)=1-x_1^2-x_2^2$,在限制条件$g(x_1,x_2)=x_1+x_2-1=0$下的驻点:<p>
$L(x,\lambda)=1-x_1^2-x_2^2+\lambda(x_1+x_2-1) \tag{E.5}$
分别求导有:<p>
    $-2x_1+\lambda=0 \tag{E.6}$
    $-2x_2+\lambda=0 \tag{E.7}$
    $x_1+x_2-1=0 \tag{E.8}$
所以$x_1^*=0.5,x_2^*=0.5,\lambda=1$

不等式限制：$g(x)\geq0$,两种情况：
- g(x)>0,限制条件未激活，$\lambda=0$
- g(x)=0，限制条件激活，$\lambda>0$

```python
# 书中图

fig, ax = plt.subplots(figsize=(10, 10))
d = plt.imread('./graph/不等式限制.png')
plt.axis('off')
ax.imshow(d)
```

综合两种情况$\lambda g(x)=0$.限制条件为:<p>
    $g(x)\geq 0 \tag{E.9}$
    $\lambda \geq 0 \tag{E.10}$
    $\lambda g(x)=0 \tag{E.11}$<p>
 叫做KKT条件。
    求极小值变为<p>
        $L(x,\lambda)=f(x)-\lambda g(x)$

结合等式和不等式限制：<p>
$$
L\left(\boldsymbol{x},\left\{\lambda_{j}\right\},\left\{\mu_{k}\right\}\right)=f(\boldsymbol{x})+\sum_{j=1}^{J} \lambda_{j} g_{j}(\boldsymbol{x})+\sum_{k=1}^{K} \mu_{k} h_{k}(\boldsymbol{x}) \tag{E.12}
$$<p>
    限制条件是$\mu_k \geq 0$且$\mu_kh_k(x)=0,k=1,...,K$
